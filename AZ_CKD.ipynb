```python
%%read_sql

select
    *
from
    RWD_DB.RWD.ICD_GROUPER
where
    level_4_short_description_icd10 ilike '%chronic%'
    and level_4_short_description_icd10 ilike '%kidney%'
    and level_4_short_description_icd10 ilike '%disease%'
```

```python
%%read_sql
create or replace table st_ckd1 as
select
    patient_id,
    left(year_of_service, 4) as year,
    claim_number,
    diagnosis,
    diagnosis_sequence
    
from
    RWD_DB.RWD.RAVEN_CLAIMS_SUBMITS_DIAGNOSIS
where
    diagnosis in ('N183')
```

```python
%%read_sql
select count (distinct patient_id)
from
    st_ckd1
    
where 
    left(patient_id, 5) != 'XXX -'
```
