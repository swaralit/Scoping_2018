# Pelican: using SNOMED code
Looking for ICD-10 I20.9 and ICD-9 4139

## SNOMED table

```python
snow.select (" select * from RWD_DB.RWD.PELICAN_DIAGNOSIS_SNOMED limit 3")
```

```python
%%read_sql
create or replace table st_angina as
select
    *
from
    RWD_DB.RWD.PELICAN_DIAGNOSIS_SNOMED
where
    concept_id in ('315025001')
```

```python
snow.select ("select count (distinct diagnosis_id) from st_angina")
```

```python
snow.select (" select * from RWD_DB.RWD.PELICAN_DIAGNOSIS limit 3")
```

## Diagnosis table

```python
%%read_sql
create or replace table st_angina2 as
select
    diagnosis_id, patient_id, start_date 
from
    RWD_DB.RWD.PELICAN_DIAGNOSIS
```

## Joining SNOMED and diagnosis tables

```python
%%read_sql
create or replace table st_ra1 as
    select a.*,
           b.patient_id, start_date
from 
    st_angina a
inner join 
    st_angina2 b
on a.diagnosis_id=b.diagnosis_id
```

```python
snow.select ("select count (distinct patient_id) from st_ra1")
```

# Albatross: using SNOMED code

```python
snow.select (" select * from RWD_DB.RWD.ALBATROSS_EHR_PROBLEMS limit 3")
```

```python
%%read_sql
create or replace table st_alb_angina as
select
    snomed,
    name,
    category,
    left(recordeddttm, 4) as recorded_year,
    left(encrypted_key_1, 8)||left(encrypted_key_2, 8) as pt_id
from
    RWD_DB.RWD.ALBATROSS_EHR_PROBLEMS
    where snomed in ('315025001')
```

```python
snow.select ("select count (distinct pt_id) from st_alb_angina")
```

```python
snow.select ("select * from st_alb_angina limit 100")
```
